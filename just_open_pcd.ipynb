```python
import glob 
import open3d as o3d
import time

path = '/home/shashank/Documents/UniBonn/Sem2/MSR_P04/DataSets/custom_dataset/sequence1'
pcd_files = glob.glob(path + "/*.pcd")
pcd_files.sort()

# Create a visualization window
vis = o3d.visualization.Visualizer()
vis.create_window()

# Create an empty list to hold the point clouds
point_clouds = []
pcd = o3d.io.read_point_cloud(pcd_files[0])
vis.add_geometry(pcd)
for pcd_file in pcd_files:
    print("Loading:", pcd_file)
    temp = o3d.io.read_point_cloud(pcd_file)
    pcd.points = temp.points
    if pcd is None:
        print("Failed to load:", pcd_file)
        continue
    
    # Add the point cloud to the visualization
    
    # point_clouds.append(pcd)
    
    # Update the visualization
    vis.update_geometry(pcd)
    vis.poll_events()
    vis.update_renderer()

    # Add key to pause
    vis.run()
    time.sleep(0.1)
    
    
    # Pause for a short duration before updating with the next point cloud

# Close the visualization window when done
vis.destroy_window()
```
